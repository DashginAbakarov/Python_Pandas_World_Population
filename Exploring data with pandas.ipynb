```python
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
%pip install seaborn
import seaborn as sns
```

```python
# we read our world population data with pandas
world_p = pd.read_csv('world_population.csv')
```

```python
# when we display our dataset it is always good to set some display options in advance.
# In this case we have adjust our float number 1 digit after decimal point when we display dataset.
# it is totally worth to do it because it helps to avoid complexity and easy to read analyse the data
pd.set_option('display.float_format', lambda x: '%.1f' %x)
```

```python
# To look at statistical onformation of data set:
world_p.describe()
```

```python
#To check how many null elements we have for each column, we use isnull() and sum(). 
#The isnull() method returns True or False for each element, indicating whether it is null or not. 
#When combined with sum(), it counts the number of True values (i.e., the number of null elements) in each column.
world_p.isnull().sum()
```

```python
# we also need to check how many unique values we have for each column.
# in our case everything seems totally alright. Rank, CCA3, Country, capital all contain 234 unique values.
# this means they are match to number of rows. Rest of them also are logical
world_p.nunique()
```

```python
# we start analysing our data just sorting them according to the population for 2020 with ascending order. 
# We also want to display only first 5 columns
world_p.sort_values(by='2020 Population', ascending=False).head()
```

```python
# As a next step we have done correlation between different columns. But first we need to select only column with valuable information
world_p.iloc[:, 5:].corr()
# I see that the perfect positive correlation between population counts for different years shows that 
#population numbers are very steady and grow in a stable way over time. 
# There is also moderate positive correlation between the size of an area and its population. 
#This means bigger countries or regions usually have more people, but it's not always true.
#I notice that the near-zero correlation between population density and population counts means that 
# How crowded an area is doesn't depend on how many people live there. So, areas with a lot of people aren't always crowded.
#I see that when there are more people, the growth rate slows down a bit. 
```

```python
# The same correlation analysis can be done using heat map as well. Because it is more interactive and easy to understand
sns.heatmap(world_p.iloc[:, 5:].corr(), annot=True)
plt.rcParams['figure.figsize'] = (20,8)
plt.show()
```

```python
# we can calculate total population for continent
# to do this we need to use two steps.
# Firstly we grouped the values according to the continent becasue population categorised according not only for continent
#but also Rank, CCA3, Country and Capital.
# second step we calculate the sum
world_p.columns
```

```python
# Here we grouped according to Continent
a = world_p.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']]
a
```

```python
# here we calculated population sum value for all years
new_pop = a.sum(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```

```python
# here sum
new_pop
# we have reset index to make sure the Continent is the column name
new_pop = new_pop.reset_index()
```

```python
# we have renamed newly added index column to Continent  
new_pop = new_pop.rename(columns={'index': 'Continent'})
```

```python
new_pop
```

```python
# Plot the population data
plt.figure(figsize=(12, 6))
# we create a list of years
#as we know names of columns are the years we need therefore we select them excep continent and make a list
years = new_pop.columns[1:].tolist()
# we make a loop in the years to plot population through the years
for year in years:
    plt.plot(new_pop['Continent'], new_pop[year], marker='o', linestyle='-', label=year)
# Adding labels, title, grid, and legend
plt.xlabel('Continent', fontsize=12)
plt.ylabel('Number of Population', fontsize=12)
plt.title('Population over the Years by Continent', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Year')
# rotating label can help to read well
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#now we want to put x axis years y axis number of population in order to make more readably grapgh
#we trasnpose our data
new_pop_1 = new_pop.T
new_pop_1
# issue is that continet become the row of dataframe
```

```python
#to solve this issue first we set first rows as a dataframe headers
new_pop_1.columns = new_pop_1.iloc[0]
new_pop_1
```

```python
#now we remove first row from the dataframe
new_pop_1 =new_pop_1[1:]
new_pop_1
```

```python
#to remove population from index we should reset index
new_pop_1 = new_pop_1.reset_index()
# and also we change name of index to Year for logical approach
new_pop_1=new_pop_1.rename(columns={'index' : 'Year'})
new_pop_1
```

```python
#we have small issue related to the index, though we can continue without fix it but to make our dataframe visually more clear
#rename the axis one with None to remove Continent
new_pop_1 = new_pop_1.rename_axis(None, axis=1)
new_pop_1
```

```python
# we create a list of years
#as we know names of columns are the years we need therefore we select them excep continent and make a list as we did before
years = new_pop_1.columns[1:].tolist()
for continent in years:
    plt.plot(new_pop_1['Year'], new_pop_1[continent], marker='o', label=continent)  
# Adding labels, title, grid, and legend
plt.xlabel('Year', fontsize=18)
plt.ylabel('Number of Population', fontsize=18)
plt.title('Population over the Years by Continent', fontsize=18)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Continent', fontsize=14)
# rotating label can help to read well
plt.xticks(rotation=45,fontsize=15)
plt.tight_layout()
plt.show()
```
